```python
import numpy as np
import pandas as pd
```

```python
from tqdm.notebook import tqdm
from PIL import Image
import os
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import RepeatVector
from keras.layers import TimeDistributed
from keras.utils import plot_model
```

### read extracted features

```python
ex_feat = np.load("extract_features.npy")
```

```python
step = 5

def sliding_window(datas,steps=1,width=step):
    win_set=[]
    for i in tqdm(np.arange(0,len(datas),steps)):
        temp=datas[i:i+width]
        if temp.shape[0] == width:
            win_set.append(temp)
    return np.array(win_set)
```

```python
data_input = sliding_window(ex_feat,steps=5,width=step)
```

```python
data_input.shape
```

### define model

```python
model = Sequential()
model.add(LSTM(200, activation='relu', input_shape=(5,32)))
model.add(RepeatVector(5))
model.add(LSTM(200, activation='relu', return_sequences=True))
model.add(TimeDistributed(Dense(32)))
model.compile(optimizer='adam', loss='mse')
```

### fit the model

```python
model.fit(data_input, data_input, epochs=1000)
```

```python
model.predict(data_input[0:1])
```
